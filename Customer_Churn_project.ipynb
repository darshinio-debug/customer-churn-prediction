Random Forest:

Build a Random Forest model where dependent variable is ‘Churn’ and independent variables are ‘tenure’ and ‘MonthlyCharges’:

a. Divide the dataset in 70:30 ratio

b. Build the model on train set and predict the values on test set

c. Build the confusion matrix and calculate the accuracy

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import *
```

```python
x=df[['MonthlyCharges', 'tenure']]
y=df['Churn']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.30, random_state= 42)
```

```python
rf= RandomForestClassifier(max_depth=5, n_estimators=100)
```

```python
rf.fit(x_train,y_train)
```

```python
import pandas as pd

importance = rf.feature_importances_
features = x.columns

feat_imp = pd.Series(importance, index=features).sort_values(ascending=False)

feat_imp.head(10).plot(kind='bar')
plt.title('Feature Importance')
plt.savefig('feature_importance.png')
plt.show()
```

```python
import pandas as pd


features = x.columns

feat_imp = pd.Series(importance, index=features).sort_values(ascending=False)

feat_imp.head(10).plot(kind='bar')
plt.title('Feature Importance')
plt.savefig('feature_importance.png')
plt.show()
```

```python
y_pred =rf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
param_grid={
    'n_estimators':[100,200,300],
    'max_depth':[None, 5,10,15],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4],
    'criterion':['gini','entropy']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search= GridSearchCV(estimator=rf, param_grid= param_grid, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
from sklearn.metrics import roc_curve, auc

y_pred_prob = rf.predict_proba(x_test)[:, 1] # Get probabilities for the positive class (Churn=1)
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.plot([0,1], [0,1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.savefig('roc_curve.png')
plt.show()
```
